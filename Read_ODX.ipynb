```python
with open('СГТ.odx') as file:
    content = file.read()
pattern=r'#Y-Values\n(.*?)\n#RefSpeed'
text = re.search(pattern, content, re.DOTALL).group(1)
x=text.split(' ')
#print(x)
type(x)
data=[float(i) for i in x]
#print(data)
file.close()
```

```python
frequency=np.arange(0, 1000, 0.625).tolist()
len(frequency)
len(data)
```

```python
plt.figure(figsize=(30, 10))
from matplotlib import pyplot as plt
plt.plot(frequency,data, color='red')
plt.xlabel('frequency, Hz', fontsize=14)
plt.ylabel('Speed', fontsize=14)
plt.grid(True)
plt.show()
```
